```python
import matplotlib.pyplot as plt

import numpy as np
import pandas as pd
import seaborn as sns
```

```python
np.random.seed(0)

df = pd.DataFrame(data={'a':np.random.randint(0, 100, 30),
                        'b':np.random.randint(0, 100, 30),
                        'c':np.random.randint(0, 100, 30)})
df.head()


plt.figure(figsize=(15, 8))

plt.subplot(2, 2, 1)
plt.plot(range(30), df['a'])

plt.subplot(2, 2, 2)
plt.plot(range(30), df['b'])

plt.subplot(2, 2, 3)
plt.plot(range(30), df['c'])

plt.subplot(2, 2, 4)
plt.plot(range(30), range(30))

plt.show()
```
