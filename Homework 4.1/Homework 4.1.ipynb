```
from bs4 import BeautifulSoup
import requests 
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}

req = requests.get('https://www.pravda.com.ua/rus/news/')
soup = BeautifulSoup(req.text, 'lxml')

jsonData = []

for news in soup.find('div', class_ = 'container_sub_news_list_wrapper mode1').find_all('div', class_ = "article_news_list"):
    jsonData.append({"Час" : news.find('div', class_="article_time").text, "Новина" : news.find('div', class_="article_header").text, "Посилання" : news.a.get('href')})

jsonData
```
